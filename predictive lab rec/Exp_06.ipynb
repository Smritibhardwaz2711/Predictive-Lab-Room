```python
import pandas as pd
iris = pd.read_csv("D:\\data sets\\IRIS (1).csv")   # Load dataset
print(iris)

import numpy as np
x = iris[['SepalLengthCm', 'SepalWidthCm']]         # Independent variables
y = iris.PetalLengthCm                              # Dependent variable

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)  # Split data

from sklearn.linear_model import LinearRegression
LR = LinearRegression()
training = LR.fit(x_train, y_train)                 # Train model
y_pred = LR.predict(x_test)                         # Predict on test data
print(y_pred)                                       # Print predictions
```
